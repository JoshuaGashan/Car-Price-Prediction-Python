# find outliers

```python
# find the IQR of engine_displacement(engine_displacement is the measure of the cylinder volume swept by all of the pistons of a piston engine, excluding the combustion chambers.)
```

```python
q1 = df['engine_displacement'].quantile(0.25)
q3 = df['engine_displacement'].quantile(0.75)

iqr = q3 - q1

print("Interquartile Range (IQR):", iqr)
```

```python
# find the IQR Of  (The legroom in inches for the passenger seat)
```

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)

iqr = q3 - q1

print("Interquartile Range (IQR):", iqr)
```

```python
q1 = df['year'].quantile(0.25)
q3 = df['year'].quantile(0.75)

iqr = q3 - q1

print("Interquartile Range (IQR):", iqr)
```
